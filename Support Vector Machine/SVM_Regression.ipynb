```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/python/Dataset/Used_Bikes.csv")
df
```

```python
new_df=df.drop(['city','bike_name'],axis='columns')
new_df
```

```python
df['brand'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
new_df['owner']=le.fit_transform(new_df['owner'])
```

```python
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
new_df['brand']=le.fit_transform(new_df['brand'])
```

```python
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
x=new_df.drop(['price'],axis='columns')
y=new_df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
model=SVR()
model.fit(x_train,y_train)
```

```python
y_predicted=model.predict(x_test)
```
