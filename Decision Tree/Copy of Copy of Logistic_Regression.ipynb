```python
#lets define the sigmoid function now and do math with hand
import math
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
def prediction_function(age):
  z=0.042*age-1.53 #0.04150133 ~0.042 and -1.52726963 ~ -1.53
  y=sigmoid(z)
  return y
```
