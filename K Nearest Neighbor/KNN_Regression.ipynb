```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/python/Dataset/travel_packages.csv")
df
```

```python
new_df=df.drop(['index','about_trip','emi'],axis='columns')
new_df
```

```python
new_df['time'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
new_df['time']=le.fit_transform(new_df['time'])
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
new_df['price']=pd.Series(new_df['price']).str.replace(',','').astype(int)
```

```python
new_df
```

```python
new_df['place']=le.fit_transform(new_df['place'])
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(class_mapping)
```

```python
x=new_df.drop(['price'],axis='columns')
y=new_df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
model=KNeighborsRegressor(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
y_predicted= model.predict(x_test)
```
