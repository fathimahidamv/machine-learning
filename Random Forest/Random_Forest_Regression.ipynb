```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestRegressor
%matplotlib inline
```

```python
df=pd.read_csv("/content/drive/MyDrive/python/Dataset/hiring (1).csv")
df.head()
```

```python
new_df=df.drop(['experience'],axis='columns')
new_df
```

```python
new_df['test_score']=new_df['test_score'].fillna(new_df['test_score'].median())
new_df
```

```python
#independent variable--input
x=new_df.drop(['salary'],axis='columns')
#dependent variable--output
y=new_df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python

model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
x_test
```

```python
x_train
```

```python
y_predicted=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
